- push

```python
# 파이썬 리스트에서 stack구조의 push/pop과 동일한 메서드가 있다.

stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print(stack)
```

- pop

```python
num = stack.pop()   # 30
print(num)
print(stack)

num = stack.pop()   # 20
print(num)
print(stack)

num = stack.pop()   # 10
print(num)
print(stack)
```

- 웹 브라우저 "뒤로가기" 구현

```python
stack = []

stack.append("네이버")
stack.append("구글")
stack.append("ChatGPT")

print("현재 방문 기록 :", stack)
print("뒤로가기 :", stack.pop())    
print("현재 페이지 :", stack[-1])   # pop연산 후 top 연산
print("뒤로가기 :", stack.pop())    
print("현재 페이지 :", stack[-1])
```

- "접시(동전) 쌓기" 구현 

```python
stack = []

stack.append("접시1")
stack.append("접시2")
stack.append("접시3")

while len(stack) > 0:
    print(stack.pop())
```

[과제] 100엔 스시 집 계산
1) 자료 구조 [{}, {}] 의 stack구현

```python
plates = []

plates.append({"color":"red", "price":100})
plates.append({"color":"blue", "price":150})
plates.append({"color":"blue", "price":150})
plates.append({"color":"black", "price":200})
plates.append({"color":"red", "price":100})
plates.append({"color":"black", "price":200})
plates.append({"color":"black", "price":200})

print(plates)

total_price = 0

while len(plates) > 0:
    p = plates.pop()
    total_price += p["price"]

print(f"¥{total_price:,}엔")
print(plates)
```
