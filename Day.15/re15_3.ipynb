## 1. 문자열에서 숫자를 모두 찾아 출력하시오.

# ['y', 't', 'h', 'o', 'n']

# 출력
# ['123', '456']

```python
import re

text = "Python123AI456"
result = re.findall(r"\d+", text)

print(result)
```

## 2. 문자열에서 소문자만 모두 출력하시오.

text = "PythonAI"

# 출력
# ['y', 't', 'h', 'o', 'n']

```python
text = "PythonAI"
result = re.findall(r"[a-z]", text)

print(result)
```

## 3. 문자열에 숫자가 포함되어 있는지 검사하시오.

text = "Python123"

```python
text = "Python123"

if re.search(r"[0-9]", text):
    print("숫자가 있습니다.")
```

## 4. 문자열에서 전화번호를 찾아 출력하시오.

text = "문의 : 010-1234-5678"

# 출력
# 010-1234-5678

```python
text = "문의 : 010-1234-5678"

result = re.search("\d{3}-\d{4}-\d{4}", text)

print(result)
```

## 6. 문자열에서 숫자를 모두 *로 변경하시오.

text = "A10 B20 C30"

# 출력
# A** B** C**

```python
text = "A10 B20 C30"

result = re.sub(r"\d", "*", text)

print(result)
```

7. 문자열에서 HTML태그("<"로 시작하고 ">"로 끝나는 문자)를 제거하시오.

html = "<h1>Python</h1>"

# 출력
# Python

```python
html = "<h1>Python</h1>"

result = re.sub(r"[^>]+*", html)

print(result)
```
