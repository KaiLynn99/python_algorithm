- re.findall() 모든 검색 결과를 리스트로 반환한다.

```python
import re

text = "apple banana orange"
result = re.findall(r"an", text)
print(result) # ['a', 'a', 'a', 'a', 'a']
```

```python
import re

text = "A10 B20 C30"
numbers = re.findall(r"\d+", text) # \d 숫자한문자, \d+ 연속된 숫자 한문자이상
print(numbers)
```

실습1) 상품명에서 가격만 출력

```python
import re

products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]

for product in products:
    price = re.search(r"\d+", product)
    print(price.group())
```

```python
import re

products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]

price_list = []

for product in products:
    name = re.search(r"[가-힣]+", product)
    price = re.search(r"\d+", product)
    # print(price.group())
    price_list.append({name.group():int(price.group())})

print(price_list)
# [{사과 : '3500'}, {배 : '4200'}, {수박 : '18000'}]
```

## 실습3) 이메일 검증

```python
import re

emails = [
    "python@네이버.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")
```

## 실습4) URL 검증

```python
import re

urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"
# ^https? - "http" 이거나 "https" 로 시작
# "://" - ""http:/"/" 이거나 "https://" 로 시작
# [\w-]+ - ["\w" 이거나 "-"]+  ex)http://abc
# ([\w-])+\. - ex)http://abc.abc.
# (/[^\s]*)? - ex)http://abc.abc.com
# http://abc.abc.com/

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

## 실습5) 주민등록변호 형식 검사(외국사람 제외)

```python
import re

nums = [
    "901010-1234567",
    "000101-4234567",
    "000101-7234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^(\d{6})-([1-4]\d{6})$"

for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
