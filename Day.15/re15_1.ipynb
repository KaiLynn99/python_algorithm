## 4) 문자제외 [^ ] 

- 대괄호 안에 ^(캐럿 문자 caret)는 괄호안에 문자를 제외한 문자

- 대괄호 `안` `밖`에 따라 기능이 다르다.

```python
import re

text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
import re

text = "b7bc123"
result = re.findall(r"^[0-9]", text) 
print(result)
```

## 5) 아무 문자 .

- . 은 줄바꿈을 제외한 아무 문자 한 글자를 의미한다.

```python
import re

text = "cat ct cot cut coot"
result = re.findall(r"c.t", text) # "c_t" 이런 패턴의 문자(c로 시작, t로 끝나는 문자열길이 3) 찾기
print(result)
```

## 6) 시작(^)과 끝($)

- 문자열 안에서 패턴의 문자를 찾기보다는 문자열 전체가 특정 문자열로 시작 or 끝나는 지를 확인할때 사용

- ^가 [] 밖에 있으면 → 문자열의 시작 부분만

- ^가 [] 안에 있으면 → ~이 아닌(Not)

```python
import re

# findall함수에서는 ^패턴을 사용해도 의미있는 반환값을 얻을 수 없다(^패턴은 전체 문자열을 하나로 다루기 때문)
text = "Python Python python"
print(re.findall(r"^Py", text)) # 영어 대문자 Py

text = "python Python python"
print(re.findall(r"^Py", text)) # 영어 소문자 py

text = "Python Python python"
print(re.findall(r"(Py)", text)) # 가독성 위해 ()로 감싸기

print("_"*10)
text = ["Python Python python", "python Python python"]

for t in text:
    print(re.findall(r"^Py", t))
```

- search(정규표현식, text) 함수를 사용 -> 중간에서 찾을 때

- 문자로 시작하면 match객체, 아니면 None 반환 -> 처음부터 찾을 때

```python
import re

text = "Python Python python"
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.search(r"Py", text))
print(re.search(r"^py", text)) 
print(re.search(r"py", text)) 
```

```python
import re

text = "Python"

print(re.search(r"hon$", text)) 
print(re.search(r"abc$", text)) 
```

## 7) 수량자 * 와 +

* 앞에 문자는 0번이상 여러번 반복될 수 있다. 최소 개수 0개

+ 앞에 문자는 1번이상 여러번 반복될 수 있다. 최소 개수 1개

```python
import re

text = "Pyyyyython"

# 시작이 Py 찾기
print(re.search(r"^Py", text)) 
# 시작이 Pyyyyy.... 찾기
print(re.search(r"^Py*", text)) 
```

```python
# 시작이 Py 찾기
print(re.search(r"Py", text)) 
# 시작이 Pyyyyy.... 찾기
print(re.search(r"Py*", text)) 
print(re.search(r"Py+", text)) 
```

- match()함수 : 문자열 처음부터 패턴이 맞는지만 검사

- fullmatch()함수 : 문자열 전체가 패턴과 일치해야 함 

```python
import re

texts = ["3", "123", "", "abc", "12a"]

for t in texts:
    print(t, "->", bool(re.fullmatch(r"\d+", t))) # \d* : \d+ : 전체에서 숫자가 1개 이상 있어야 한다.
```

```python
import re

texts = ["3", "123", "", "abc", "12a"]

for t in texts:

    print(t, "->", bool(re.fullmatch(r"\d*", t))) 
```

## 8) 수량자 ? 

- 앞 문자가 0개 또는 1개 존재한다.

- "ab?c" : b가 없거나 1개 있거나 -> "abc"(O) 와 "ac"(O)는 매칭,  "abbc"(X) b가 2개라서 매칭 안됨

```python
import re

words = [
    "ae",
    "abc",
    "abbcd",
    "abbbcde",
    "acde"
]

# ab?c : b문자가 0개 or 1개 패턴

for word in words:
    print(re.match(r"ab?c", word))
```

## 10) 수량자 {n}

- 앞 문자가 정확히 n번 반복

```python
import re

text = "1,23,456,7890,77a8888"
print(re.findall(r"\d{3}", text)) # 숫자문자가 3번연속 나오는 구간만 반환 리스트에 추가
```

```python
import re

text = "1,23,456,7890,77a8888"
print(re.search(r"\d{3}", text)) 
```

## 12) OR( | ) 연산

여러 패턴 중 하나와 일치하면 된다.

```python
import re

text = "cat, dog, bird, cats, dog"

print(re.findall(r"cat|dog", text)) 
print(re.findall(r"[(cat)(dog)]", text)) 
```

## 13) group()함수 매칭된 결과물을 반환

```python
import re

phones = [
    "02-1234-5678",
    "010-1234-5678",
    "010-123-5678",
    "010-1234-11111"
]

# "숫자3시작-숫자4-숫자4끝나는" 찾기
for p in phones:
    result = re.search(r"^\d{3}-\d{4}-\d{4}$", p)
    if result != None:
        print(result)
        print(result.group())
```
