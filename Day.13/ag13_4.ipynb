# 실습 과제 1. 누적합(Prefix Sum)

```python
steps = [5231, 6120, 4890, 7005, 6500, 7200, 6800]

prifix = [0]

for n in steps:
    prifix.append(prifix[-1] + n)

for i in range(len(steps)):
    print(f"{i+1}일 누적 : ", prifix[i+1])
```

# 실습 과제 2. 누적합(Prefix Sum) - 1번 응용

```python
steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]

steps_count = []
prefix = [0]
dates = []

for day in steps:
    for date, count in day.items():
        dates.append(date)
        steps_count.append(count)

print(dates)

for n in steps_count:
    prefix.append(prefix[-1] + n)

for i in range(len(steps)):
    print(f"{dates[i]} 누적 : ", prefix[i+1])
```
