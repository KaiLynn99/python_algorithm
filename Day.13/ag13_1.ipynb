```python
nums = [1, 3, 5, 7, 9, 10, 20, 10]

# 1번째 부터 8번째 까지
# 인덱스 0부터 7

t1 = 0
for i in range(8):
    t1 = t1 + nums[i]
print(t1)


# 1번째 부터 3번째 까지
t2 = 0
for i in range(3):
    t2 = t2 + nums[i]
print(t2)
```

```python
nums = [1, 3, 5, 7, 9, 10, 20, 10]
# 누적합 0 부터 시작
prifix = [0]

for n in nums:
    # 0 + 1
    # prifix[1] + 3
    # prifix[2] + 5
    ...
    prifix.append(prifix[-1] + n)

print(nums)
print(prifix)
```

```python
# 1 처음부터 8 끝까지 인덱스(0,7)
prifix[8] - prifix[1-1]
```

```python
# 1부터 7까지(0,6)
prifix[7]
```

```python
# 원본의 5번째에서 7번째까지 (인덱스4,6)
s = 5
e = 7

prifix[e] - prifix[s-1]
```
