1) 선형 탐색 - 반복 순환

```python
numbers = [15, 8, 23, 7, 31, 18]
target = 23

count = 0

for i in range(len(numbers)):
    count += 1
    if numbers[i] == target:
        print("찾은 위치 :", i)
        break # 찾으면 남은 원소는 비교하지않고 break

print("비교 횟수 :", count)
```

2) 최대값 찾기 - 반복 순환

```python
numbers = [15, 8, 23, 7, 31, 18]

# 첫번째 값을 최대값으로 설정
max_num = numbers[0]
count = 0

for n in numbers:
    count += 1
    if n > max_num:    # n > max_num: -> 최솟값
        max_num = n

print("최댓값 :", max_num)
print("비교 횟수 :", count)
```

3) 버블 정렬 - 중첩 반복 알고리즘

```python
numbers = [5, 3, 8, 1, 4]

# 데이터 5개
# 1회전 : 4번 비교
# 2회전 : 3번 비교
# 3회전 : 2번 비교
# 4회전 : 1번 비교

count = 0

for j in range(len(numbers) - 1):
    for i in range(len(numbers) - 1 - j):
        count += 1
        if numbers[i] > numbers[i + 1]:
            numbers[i], numbers[i + 1] = numbers[i + 1], numbers[i] # 우측 패킹 -> 대입 -> 좌측 언패킹
            # temp = numbers[i]
            # numbers[i] = numbers[i+1]
            # numbers[i+1] = temp

print(numbers)
print("비교 횟수 :", count)
```

```python
import random
import time
```

```python
# 0~9999 사이의 숫자 1000개를 무작위로 생성
numbers = random.sample(range(10000), 1000) 
len(numbers)

# 실행 시간 측정 시작
start = time.time()

# 마지막 원소를 찾도록 설정(최악의 경우 : 최대 실행 횟수 확인)
target = numbers[-1]
count = 0

# 처음부터 끝까지 하나씩 비교
for n in numbers:
    count += 1
    if n == target:
        break

# 실행 시간 측정 종료
end = time.time()

print("선형 탐색 :", end - start, "count:", count)
```
