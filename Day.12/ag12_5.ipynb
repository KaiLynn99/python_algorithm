- 오른쪽 데이터 추가 append

```python
from collections import deque

dq = deque()
dq.append(10)
dq.append(20)
dq.append(30)

print(dq)
```

- 왼쪽 데이터 추가 appendleft

```python
dq = deque()
dq.append(20)
dq.append(30)
dq.appendleft(10) # 0번째 위치에 삽입

x = dq.pop() # 스텍처럼

print(x)
print(dq)

print(dq.popleft()) # 큐 처럼
print(dq)
```

- 오른쪽 삭제 pop

```python
from collections import deque

dq = deque([10, 20, 30])
x = dq.pop()

print(x)
print(dq)
```

```python
# 스텍 필요 -> 효율성 Deque X -> list 사용
# 큐 필요 -> 효율성 O -> Deque 사용
# 덱 필요 -> Deque 사용
# 자료구조의 회전 -> 시계방향? 반시계? 양방향 삽입/삭제 가능 > Deque 사용
```

- 회전 roate

```python
from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전) / -1 : 반시계

print(dq) # deque([5, 1, 2, 3, 4])

dql = deque([1, 2, 3, 4, 5])
x = dql.pop()
dql.appendleft(x)
print(dql)
```

- 최근 방분 기록 n개 저장

```python
from collections import deque


history = deque(maxlen=5) # maxlen=n n개만 저장

history.append("google")
history.append("naver")
history.append("youtube")
history.append("github")
history.append("chatgpt")

print(history)

history.append("python")

print(history)
```

- 카드 게임(원카드, 우노, ..) 시뮬레이션

```python
from collections import deque


cards = deque([1, 2, 3, 4, 5]) # 처음 받은 카드 5장

while len(cards) > 1:
    print("버림 :", cards.popleft()) # 1번 카드를 버림
    cards.rotate(-1)
    # cards.append(cards.popleft())
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
