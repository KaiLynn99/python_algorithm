- 데이터 넣기(Enqueue)

```python
queue = []

queue.append("A")
queue.append("B")
queue.append("C")

print(queue)
```

- 데이터 꺼내기(Dequeue)

```python
queue = ["A", "B", "C"]

print(queue.pop(0)) # ["A", "B", "C"]
print(queue)

print(queue.pop(0))
print(queue)

print(queue.pop(0))
print(queue)
```

```python
queue = ["A", "B", "C"]

print(queue[0]) # ["A", "B", "C"]
print(queue[1:] )


print(queue[1])
print(queue[2:])

print(queue[2])
print(queue[3:])
```

```python
queue = ["A", "B", "C"]

pop_index = 0

while pop_index < len(queue):
    print(queue[pop_index]) # ["A", "B", "C"]
    print(queue[pop_index + 1:])
    pop_index += 1

print(queue)
```

- 은행 번호표 예제

```python
queue = []

queue.append("홍길동")
queue.append("이순신")
queue.append("김유신")

print("대기 :", queue)
print("입장 :", queue.pop(0))
print("현재 대기 :", queue)

print("대기 :", queue)
print("입장 :", queue.pop(0))
print("현재 대기 :", queue)

print("대기 :", queue)
print("입장 :", queue.pop(0))
print("현재 대기 :", queue)
```

- 프린트 출력 큐 예제

```python
queue = []

queue.append("문서1")
queue.append("문서2")
queue.append("문서3")

while len(queue) > 0:
    print("출력 :", queue.pop(0))
    print(queue)
```

- 음식 주문 예제

```python
queue = []

queue.append("햄버거")
queue.append("피자")
queue.append("라면")

print("주문 목록 :", queue)

print("조리 시작 :", queue.pop(0))
print("조리 시작 :", queue.pop(0))

print("남은 주문 :", queue)
```

```python
# 큐 대신 덱(양방향 입출력 가능) 을 사용 -> 더 편리

from collections import deque

queue = deque()

queue.append("A")
queue.append("B")
queue.append("C")

print(queue.popleft()) # 큐
print(queue)
print(queue.pop()) # 스택
print(queue)
```
