## 1. 팀장 번호 구하기 (산술 연산자)

1번부터 100번까지 학생이 있다.

1~10번은 1팀 

11~20번은 2팀 

21~30번은 3팀 

... 

각 팀은 10명씩이며 첫 번째 학생이 팀장이다.

학생 번호를 입력받아 해당 학생의 팀장 번호를 출력하는 프로그램을 작성하시오.

실행 예)

학생 번호 : 8
팀장 번호 : 1

학생 번호 : 15
팀장 번호 : 11

```python
num = int(input("학생 번호 : "))
# num

# (num-1) // 10
# 1~10 -> 0
# 11~20 -> 10
leader_number = ((num-1) // 10 * 10) + 1

print(leader_number)
```

## 2. 홀수와 짝수 판별하기 (if~else)

정수를 입력받아 홀수인지 짝수인지 출력하는 프로그램을 작성하시오.

실행 예)

숫자 : 7
홀수

```python
try:
    num = int(input("숫자를 입력하세요"))

    if num % 2 == 0:
        print("짝수")
    else:
        print("홀수")
        
except ValueError as e:
    print("숫자만 입력하세요")
    # print(e)
```

## 3. 양수, 음수, 0 판별하기 (if~elif~else)

정수를 입력받아 양수, 음수 또는 0을 출력하는 프로그램을 작성하시오.

```python
try:

    num = int(input("숫자를 입력하세요"))

    if num > 0:
        print("양수")
    elif num < 0:
        print("음수")
    else:
        print(0)
        
except ValueError as e:
    print("숫자만 입력히세요")
```

## 4. 20대 출입 가능 여부

두 사람의 나이를 입력받는다.

두 사람 모두 20대(20~29세)이면 출입 가능을 출력하고,

그 외에는 출입 불가를 출력하는 프로그램을 작성하시오.

```python
try:

    a_age, b_age = map(int,input().split())

    a_age_group = a_age // 10
    b_age_group = b_age // 10

    if a_age_group == 2 and b_age_group == 2:
        print("출입 가능")
    else:
        print("출입 불가")

except ValueError as e:
        print("숫자만 입력하세요")
```

## 7. 윤년 구하기

2014년부터 2500년까지의 윤년을 모두 출력하는 프로그램을 작성하시오.

윤년의 조건

- 400으로 나누어 떨어지는 해 

- 또는 4로 나누어 떨어지지만 100으로는 나누어 떨어지지 않는 해

```python
for i in range(2014,2501):
    if (i % 400) == 0:
        print(i)
    elif ((i % 4) == 0) and ((i % 100 != 0)):
        print(i)
```

## 8. 소수 판별하기

정수를 입력받아 소수인지 아닌지 판별하는 프로그램을 작성하시오.

(소수는 1보다 큰 자연수 중 1과 자기 자신만으로 나누어지는 수이다.)

```python
num = int(input())

if num < 2:
    print("소수도 합성수도 아닙니다")

else:
    for i in range(2, num):
        if num % i == 0:
            print("합성수")
            break
    else:
        print("소수")

```

## 13. 세로 구구단 출력

2단부터 9단까지를 세로 방향으로 출력하는 프로그램을 작성하시오.

출력 포맷 답 부분은 2자리 숫자로 출력 ex) "2*1=2" -> "2*1= 2"

예시)

2*1= 2 3*1= 3 4*1= 4 5*1= 5 6*1= 6 7*1= 7 8*1= 8 9*1= 9 
2*2= 4 3*2= 6 4*2= 8 5*2=10 6*2=12 7*2=14 8*2=16 9*2=18 
...

```python
# 아래 코드가 i(1~9)가 9번 반복

for i in range(1,10): 
    for dan in range(2, 10):
        print(f"{dan} x {i} = {dan*i:>2}", end=" | ")
    print("", end="\n") # print()

O(n2)
```
