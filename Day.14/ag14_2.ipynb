## 2) 문자클래스 []
## (하나라도 일치하면 반환)

```python
import re

text = "apple banana cat"
result = re.findall(r"[abc]", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result)
```

```python
text = "apple banana cat"

result = re.search(r"[at]", text)
if result:
    print("a나t가 있다")
else:
    print("a나t가 없다")
```

## 3) 대괄호[] 안의 "-"는 범위를 나타낸다.

```python
import re

text = "A12B34"
print(re.findall(r"[0-9]", text))
```

#### - 영문대문자 찾기

```python
text = "Python JAVA AI"
print(re.findall(r"[a-zA-Z]", text))
```

## 4) 문자제외 [^ ]

```python
text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
# 대소문자 포함, 특수문자 포함, 숫자문자 포함

import re

password = input("비밀번호:")

result = []

if re.search(r"[가-힣]",password):
    result.append("한글을 사용할 수 없습니다")

if not re.search(r"[a-z]", password):
    result.append("소문자가 없습니다")

if not re.search(r"[A-Z]", password):
    result.append("대문자가 없습니다")

if not re.search(r"[0-9a-zA-Z가-힣]", password):
    result.append("특수문자가 없습니다")

if len(result) == 0:
    result.append("사용 가능한 비밀번호입니다.")

result
```
