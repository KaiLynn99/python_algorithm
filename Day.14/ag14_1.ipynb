3. 빈도수 계산

문제

상품별 판매 건수를 구하시오.

```python
count = {} # {"노트북" : 5, "마우스" : 5, "키보드" : 4, "모니터" : }5
# 노트북 없다-> 노트북:1 추가, 있으면 +1

for r in sales:
    key_name = r["상품"] 
    if key_name not in count:
        count[key_name] = 0
    count[key_name] = count[key_name] + 1 # else 줄인 버젼

print(count)

for k, v in count.items():
    print(f"{k}(는)은 {v}건 입니다")
```

4. 그룹별 집계(Group By)

문제

지역별 총 판매액을 구하시오.

```python
local_sum = {} 

for r in sales:
    price = r["판매액"]
    # price "", "결측" 이면
    if not isinstance(price, int): # if pass else ... -> if continue ...
        continue # 다음 반복으로 건너뛰기
    key_name = r["지역"]

    if key_name not in local_sum:
        local_sum[key_name] = 0
    
    local_sum[key_name] += price

print(local_sum)

for k,v in local_sum.items():
    print(f"{k}(는)은 {v:,}원입니다")
```

8. 이동평균(Moving Average)

판매액

- 슬라이딩 윈도우

```python
# 슬라이딩 윈도우에 사용될 리스트
prices = []

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

# prices

window = 3

# [0:3] -> [1:4] -> [2:5]

for i in range(len(prices) - (window - 1)): # range(len(price) - window + 1)
    total = sum(prices[i: i + window])
    avg = round(total / window, 3)
    print(f"{i+1}~{i+window} 평균 판매액: {avg:,}원") # 1~3 ???, 2~4 : ???
```

9. 누적평균(Running Average)

- Prefix Sum

```python
prices = []

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

# prices

prefix = [0]

for p in prices:
    prefix.append(prefix[-1] + p)

# prefix

avgs = []

for i in range(1, len(prefix)): # 맨 처음 0 제거
    avgs.append(prefix[i] / i)

avgs
```

11. 결측치 처리

"","결측"을 평균이나 중앙값으로 변경.

- try 예외처리 또는 if 조건분기

```python
# avg = sum(prices) / len(prices) # 평균

# 중앙값 정렬
# print(len(prices)) # 짝수

prices.sort()
print(prices)

median = 0 # 중앙값 초기화

# 짝홀에 따라 중앙값 처리 방법이 다름
if len(prices) % 2 == 1: # 홀수 이면
    median = prices[len(prices) // 2]

else: # 짝수 이면
    median = (prices[len(prices) // 2] + prices[len(prices) // 2 - 1]) / 2  
    
print(median)

print("원본", sales)

for r in sales:
    if not isinstance(r["판매액"], int):
        r["판매액"] = median

print("결측치->중앙값", sales)
```

12. 범주별 통계

```python
result = {}
for r in sales:
    p_name = r["상품"]
    price = r["판매액"]
    total_cnt = r["수량"]

    if p_name not in result:
        result[p_name] = {"합계":0, "건수":0, "최대":price, "최소":price, "총 수량":0}
        # {
        # {"키보드" :{"합계":0, "건수":0, "최대":0, "최소":0, "총 수량":0}}
        # {"모니터" :{"합계":0, "건수":0, "최대":0, "최소":0, "총 수량":0}}
        # }

    info = result[p_name] 
    info["합계"] += price    
    info["건수"] += 1
    info["총 수량"] += total_cnt

    # 최대/최소
    if info["최대"] < price:
        info["최대"] = price

    if info["최소"] > price:
        info["최소"] = price

print(result)
```
