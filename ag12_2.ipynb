1) 공간복잡도 O(1)

```python
n = int(input("숫자 입력 : "))
result = n * 2
print(result)
```

2) 공간복잡도 O(n)

```python
n = int(input("숫자 개수 : "))

numbers = []

for i in range(n):
    numbers.append(i)

print(numbers)

# 입력이 커질수록 리스트도 커진다.
```

3) 공간복잡도 O(n²)

```python
n = int(input("크기 : "))

matrix = [] # 3-> 3*3

for i in range(n):
    row = []
    for j in range(n):
        row.append(0)
    matrix.append(row)

print(matrix)
```
