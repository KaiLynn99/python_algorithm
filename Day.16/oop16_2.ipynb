```python
class Student:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)
    def __init__(self, name):
        self.name = name
    def greet(self): # 새로 추가
        print(f"{self.name}님 hello")
```

```python
# 상속하면
class HighStudent(Student):
    def study(self):
        print("공부한다.")
```

```python
# 사용
s = HighStudent("홍길동")
s.print_name()
s.study()
s.greet()
```

## 1) 딕셔너리 함수

```python
def average(student):
    return (student["kor"] + student["eng"]) / 2


student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85,
    "math": 55,
    "average": average
}

print(student["average"](student))


student2 = {
    "name": "김철수",
    "kor": 70,
    "eng": 75,
    "average": average
}

print(student["average"](student2))
```

## 2) 클래스 메서드

```python
class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = 0

    def average(self):
        return (self.kor + self.eng + self.math) / 3


s = Student("홍길동", 90, 85)
s.math = 88

s2 = Student("김철수", 70, 75)
s2.math = 78

print(s.average())
print(s2.average())
```
