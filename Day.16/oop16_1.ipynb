```python
# 클래스는 사용자 정의 컬렉션 타입 

class Student: # 대문자로 시작
    """
    문자열값
    """
    # 생성자 메서드
    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def total(self):
        return self.kor + self.eng

    def average(self):
        return self.total() / 2

    def print_info(self):
        print(f"이름 : {self.name}")
        print(f"국어 : {self.kor}")
        print(f"영어 : {self.eng}")
        print(f"총점 : {self.total()}")
        print(f"평균 : {self.average()}")
```

```python
s1 = Student("홍길동", 100, 90) 
# Student.__init__(self, name, kor, eng))

print(s1.name)
print(s1.kor)
print(s1.eng)

print()

s1.print_info()

print()

s2 = Student("김철수", 99, 95)
s2.print_info()
```

```python
class Account:

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, money):
        self.balance += money

    def withdraw(self, money):
        self.balance -= money

    def print_info(self):
        print(f"예금주 : {self.owner}")
        print(f"잔액 : {self.balance:,}원")
```

```python
acc = Account("홍길동", 10000)

acc.deposit(5000)
acc.withdraw(3000)

acc.print_info()
```

```python
class Car:

    def __init__(self, model):
        self.model = model
        self.speed = 0

    def accelerate(self):
        self.speed += 10

    def brake(self):
        self.speed -= 10

    def print_info(self):
        print(f"차량 : {self.model}")
        print(f"속도 : {self.speed:,}km/h")
```

```python
car = Car("소나타")

car.accelerate()
car.accelerate()
car.brake()

car.print_info()
```

```python
class TV:

    def __init__(self):
        self.power = False
        self.channel = 1

    def on(self):
        self.power = True

    def channel_up(self):
        self.channel += 1

    def print_info(self):
        print(f"전원 : {self.power}")
        print(f"채널 : {self.channel}")
```

```python
tv = TV()

tv.on()
tv.channel_up()
tv.channel_up()

tv.print_info()
```

```python
class Rectangle:

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return (self.width + self.height) * 2


r = Rectangle(10, 5)

print(r.area())
print(r.perimeter())
```
